optimizing production costs based on raw material factors, labor, etc.


```python
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 5*x + 1

def secant(x0, x1, iterations, tol=1e-6):
    for i in range(iterations):
        f_x0 = f(x0)
        f_x1 = f(x1)

        if f_x1 - f_x0 == 0:
            print("Division by zero occurred at iteration", i)
            return None

        x2 = x1 - (f_x1 * (x1 - x0)) / (f_x1 - f_x0)
        print(f"Iteration-{i+1}: x = {x2}, f(x) = {f(x2)}")

        if abs(x2 - x1) < tol:
            print("Converged, stopping iterations")
            return x2

        x0, x1 = x1, x2

    return x2

x0 = float(input("Enter the initial value x0: "))
x1 = float(input("Enter the initial value x1: "))
iterations = int(input("Enter the number of iterations: "))

root = secant(x0, x1, iterations)

x = [i / 100 for i in range(101)]
y = [f(i) for i in x]
plt.plot(x, y, label='f(x) = x^3 - 5x + 1')
plt.axhline(0, color='grey', lw=0.5)
plt.axvline(root, color='red', linestyle='--', label=f'Approximate Root: {root:.8f}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Root Approximation Using Secant Method')
plt.legend()
plt.grid(True)
plt.show()
```
