```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def sum_error(y, a0, a1, x):
    return np.sum((y - a0 - a1 * x) ** 2)

def regression(x, y):
    n = len(x)
    avg_y = np.mean(y)
    avg_x = np.mean(x)

    a1 = (n * np.sum(x * y) - np.sum(x) * np.sum(y)) / (n * np.sum(x**2) - np.sum(x)**2)
    a0 = avg_y - a1 * avg_x

    yi_avg_y_squared = (y - avg_y) ** 2
    error_squared = sum_error(y, a0, a1, x)
    return a0, a1, yi_avg_y_squared, error_squared

x = np.array([2600, 3000, 3200, 3600, 4000])
y = np.array([550000, 565000, 610000, 680000, 725000])

a0, a1, yi_avg_y_squared, error_squared = regression(x, y)

data = {
    'xi': x,
    'yi': y,
    '(yi-avg_y)^2': yi_avg_y_squared,
    '(y - a0 - a1*x)^2': error_squared
}

tabel = pd.DataFrame(data)
plt.scatter(x, y, color='green', label='Data')
plt.plot(x, a0 + a1 * x, color='black', label='Trendline')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresi Linear')
plt.legend()
plt.show()
```
